```python
import random


def hangman():
    
    hangman_words = [
    "python", "hangman", "programming", "computer", "code",
    "learning", "challenge", "algorithm", "developer", "variable",
    "function", "loop", "conditional", "debugging", "string",
    "dictionary", "list", "tuple", "module", "class",
    "object", "inheritance", "encapsulation", "polymorphism", "exception",
    "recursion", "import", "iteration", "indexing", "slice",
    "random", "while", "for", "break", "continue",
    "try", "except", "finally", "raise", "assert",
    "input", "output", "file", "open", "close",
    "append", "sort", "reverse", "len", "range",
    "print", "format", "join", "split", "replace"
]

    
    word = random.choice(hangman_words)
    wrong = 0
    stages = ["",
             "__________",
             "|         ",
             "|    |    ",
             "|    O    ",
             "|   /|\   ",
             "|   / \   ",
             "|         "
             ]
    rletters=list(word)
    board=["__"]*len(word)
    win = False
    
    #Begin the game
    
    print("Welcome to Hangman")
    while wrong < len(stages) - 1:
        print("\n")
        msg="Guess a letter: "
        char = input(msg)
        if char in rletters:
            idx = rletters.index(char)
            board[idx] = char
            rletters[idx] = '$'
        else:
            wrong += 1
            
        print("\n")
        print((" ".join(board)))
        e = wrong+1
        print("\n".join(stages[0:e]))
        print("\n")
    
        if "__" not in board:
            print("You win!")
            print("The word was: ")
            print("".join(board).upper())
            win = True
            break
    
    if not win:
        print("\n".join(stages[0:wrong]))
        print("\nYou Lose.")
        print("The word was... {}".format(word.upper()))
```

```python
hangman()
```
